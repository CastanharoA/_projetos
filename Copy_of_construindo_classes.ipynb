**Construtores**



*   São métodos (funções) de uma classe que começam com
*   __ init __ ()
*   São chamadas sempre que os objetos são instanciados.
*   Normalmente, utilizamos os construtores para iniciar as variáveis de uma classe.




```
#construtor de uma classe
class Carro:
  def __init__(self,cor='vermelho',n_portas=3):
    self.cor_do_carro=cor
    self.n_portas_do_carro=n_portas

  def get_cor_do_carro(self):
    return self.cor_do_carro

  def get_numero_de_portas(self):
    return self.n_portas_do_carro
```

```
#instanciando um objeto do tipo carro
carro_1=Carro(cor='azul')
carro_2=Carro()
carro_3=Carro(cor='prata')
```

```
#mostrando as cores de cada um dos carros
carro_1.get_cor_do_carro()
```

**O parâmetro "self"**



*   Referencia a instância corrente do objeto.
*   Utilizada para acessar os elementos que pertence a mesma classe.
*   A palavra "self" é uma convenção, não uma paravra-chave em python.



```
#construtor de uma classe
class Carro:
  def __init__(meu_self,cor='vermelho',n_portas=3):
    meu_self.cor_do_carro=cor
    meu_self.n_portas_do_carro=n_portas

  def get_cor_do_carro(meu_self):
    return meu_self.cor_do_carro

  def get_numero_de_portas(meu_self):
    return meu_self.n_portas_do_carro
```

# **Herança**

image.png



*   Permite reutilizar o código;
*   A subclasse ou classe filha, obtém todos os atributos da classe pai;
*   A classe filha pode adicionar novos atributos ou funções;
*   Para a classe filha não é necessário escrever todo o código da classe pai.



```
#sintaxe da herança em python
class ClassePai:
  '''Corpo da classe pai.'''
class ClasseFilha(ClassePai):
  '''Corpo da classe filha.'''
```

```
#definindo a classe formas 
class Formas:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    descricao = "A forma ainda não foi definida"
    autor = "Ainda não foi definido um autor para ela."
    def area(self):
        return self.x * self.y
    def get_perimetro(self):
        return 2 * self.x + 2 * self.y
    def set_descricao(self,texto):
        self.descricao = texto
    def set_nome_do_autor(self,texto):
        self.autor = texto
    def set_escala(self,escala):
        self.x = self.x * escala
        self.y = self.y * escala
```

```
#instanciando um retângulo
retangulo=Formas(5,8)
```

```
retangulo.area()
```

```
#definindo um quadrado
quadrado=Formas(3,3)
```

```
#poderia definir uma classe do tipo quadrado, pois ela é um quadrado
class Quadrado(Formas):
  def __init__(self,x):
    self.x = x
    self.y = x
```

```
#definindo um "quadrado" a partir da classe filha
quadrado=Quadrado(3)
```

```
#acessando os demais métodos
quadrado.area()
```

```
#acessando os atributos
quadrado.descricao
```

```
#adicionando novos métodos para a classe quadrado 
class Quadrado(Formas):
  def __init__(self,x):
    self.x = x
    self.y = x
  def perimetro_do_quadrado(self):
    return (self.x)*4
```

```
#definindo um quadrado a partir da classe filha
quadrado=Quadrado(10)
```

```
#utilizando um método da classe Pai
quadrado.set_nome_do_autor('Túlio Vieira')
```

```
#utilizando um método "exclusivo" da classe filha
quadrado.perimetro_do_quadrado()
```

**Sobrescrevendo métodos**

```
#adicionando novos métodos para a classe quadrado 
class Quadrado(Formas):
  def __init__(self,x):
    self.x = x
    self.y = x
  def perimetro_do_quadrado(self):
    return (self.x)*4
```



*   o __ init __() foi reescrito sobre o construtor original (do pai) 
*   Outros métodos também podem ser reescritos.



**Acessando os métodos da classe Pai**

```
#definindo a classe retangulo e a filha (quadrado)
class Retangulo:
    def __init__(self, comprimento, largura):
        self.comprimento = comprimento
        self.largura = largura

    def area(self):
        return self.comprimento * self.largura

    def perimetro(self):
        return 2 * self.comprimento + 2 * self.largura

# Quadrado como uma herança de um retangulo
class Quadrado(Retangulo):
    def __init__(self, comprimento):
        super().__init__(comprimento, comprimento)
```

```
#instanciando o objeto quadrado
quadrado=Quadrado(6)
```

```
print(quadrado.perimetro())
```

```
print(quadrado.area())
```

Também é possível acessar os métodos

```
class Cubo(Quadrado):
    def area_da_superficie(self):
        area_da_face = super().area()
        return area_da_face * 6

    def volume(self):
        area_da_face = super().area()
        return area_da_face * self.comprimento
```

```
#instanciando o objeto
cubo=Cubo(4)
```

```
cubo.area_da_superficie()
```

```
cubo.volume() #não precisei adicionar o comprimento 
```

**Verificando as heranças**

```
print(issubclass(Cubo,Retangulo))
```

```
#definindo a classe retangulo e a filha (quadrado)
class Retangulo:
    def __init__(self, comprimento, largura):
        self.comprimento = comprimento
        self.largura = largura

    def area(self):
        return self.comprimento * self.largura

    def perimetro(self):
        return 2 * self.comprimento + 2 * self.largura
```

```
# redefinindo os métodos para a classe quadrado
class Quadrado(Retangulo):
    def __init__(self, comprimento):
        super().__init__(comprimento, comprimento)
    def area(self):
      return self.comprimento * self.comprimento
    def perimetro(self):
      return self.comprimento*4
```
