```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Dados no formato "nota, horas"
dados = [
    (53, 1), (74, 3), (59, 5), (43, 8), (56, 10), (84, 11),
    (96, 14), (69, 15), (84, 15), (83, 19),   (70, 5), (83, 8), (75, 7), (95, 7)]

# Criando um DataFrame com os dados
df = pd.DataFrame(dados, columns=['Nota', 'Horas_de_Estudo'])

# Dividindo os dados em variáveis independentes (X) e dependente (y)
X = df[['Horas_de_Estudo']]
y = df['Nota']

modelo = LinearRegression() # Criando o modelo de regressão linear
modelo.fit(X, y) # Treinando o modelo
y_pred = modelo.predict(X) # Fazendo previsões
coef_angular = modelo.coef_[0] # Coeficiente angular (inclinação da reta)
intercepto = modelo.intercept_ # Intercepto (ponto onde a reta cruza o eixo y)
r2 = r2_score(y, y_pred)  # Calculando o coeficiente de determinação (R^2)
mse = mean_squared_error(y, y_pred) # Calculando o erro médio quadrático (MSE)

# Visualizando a reta de regressão
plt.scatter(X, y, label='Dados')
plt.plot(X, y_pred, color='red', linewidth=2, label='Reta de Regressão')
plt.xlabel('Horas de Estudo')
plt.ylabel('Nota')
plt.legend()
plt.show()

print(f"Coeficiente Angular (Inclinação da Reta): {coef_angular}")
print(f"Intercepto (Ponto onde a Reta Cruza o Eixo Y): {intercepto}")
print(f"Coeficiente de Determinação (R^2): {r2}")
print(f"Erro Médio Quadrático (MSE): {mse}")
```
